```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./guild.csv',engine='python',encoding='utf-8')
```

```python
df = pd.DataFrame(df)
df
```

```python
name_list = list(df['Name'])
name_Dict = {}
for i,v in enumerate(name_list):
    name_Dict[i] = v
name_Dict

user = df.rename(index=name_Dict)
user.drop(['Name','Rank','ID'],axis=1,inplace=True)
#참여여부  = > 전 날 비교해서 차이가 나면 1, 0이면 0으로 매핑
#미참여경고 = > 0될때마다 1
user['gap_Score'] = np.nan
user['Unparticipated'] = np.nan
print("------------------------분노용 3시즌---------------------------")
print("매일 PM 11:55 기여도 점수 기록")
print("미참여 경고 3회 누적 시 추방")
user
```

```python
user['gap_Score'] = user['Thu']-user['before_calc']

count=0
def unparti(score):
    global count
    if score != 0:
        return 0
    else:
        return count+1
    
user['Unparticipated'] = user['gap_Score'].apply(unparti)
user[user['Unparticipated'] != 0]
```
